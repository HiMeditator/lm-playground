```python
import json
from pathlib import Path

LANG_NUM = 3
lang = ['jp', 'en', 'zh']
narration = ['ナレーション', 'Narration', '旁白']
ch_files: list[Path] = []

folder_path = Path("./game_res/ks.scn.chapters")
for item in folder_path.iterdir():
    if item.is_file(): 
        ch_files.append(item.absolute())

print(f"{len(ch_files)} ks scn files found")
```

```python
for file in ch_files:
    with open(file, "r", encoding="utf-8") as f:
        scns = json.load(f)
    assert len(scns) > 0
    dialogs_texts = [
        f"# {scns[0]["title"][i]}\n\n"
        for i in range(LANG_NUM)
    ]
    for scn in scns:
        if "dialogs" in scn:
            dialogs = scn["dialogs"]
            for dialog in dialogs:
                name = [
                    dialog["name"][i] if dialog["name"][i] else narration[i]
                    for i in range(LANG_NUM)
                ]
                text = dialog["text"]
                for i in range(LANG_NUM):
                    dialogs_texts[i] += f"{name[i]}: {text[i].replace('\\n', '')}\n"
    for i in range(LANG_NUM):
        with open(f"./game_res/ks.scn.dialogs/{file.stem}.{lang[i]}.md", "w", encoding="utf-8") as f:
            f.write(dialogs_texts[i])
```
